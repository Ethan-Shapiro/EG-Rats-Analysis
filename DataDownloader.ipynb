```python
import pandas as pd
import riotwatcher
import os
from dotenv import load_dotenv
import json
load_dotenv()
```

```python
lol_watcher = riotwatcher.LolWatcher(os.getenv('RIOT_API_KEY'))
```

```python
filenames = [os.path.join('C9PlayerData', f) for f in os.listdir('C9PlayerData')]
```

```python
for player_file in filenames:
    # get matches from file
    all_match_data = []
    with open(player_file, 'r') as f:
        match_ids = f.readline().split(',')
        for match_id in match_ids:
            match_id = match_id.strip('"')

            match_data = lol_watcher.match.by_id(region='ASIA', match_id=match_id)
            all_match_data.append(match_data)
    with open(os.path.join('C9PlayerData', f'{player_file.split(".")[0]}.json'), 'w+') as out:
        json.dump(all_match_data, out)
```
